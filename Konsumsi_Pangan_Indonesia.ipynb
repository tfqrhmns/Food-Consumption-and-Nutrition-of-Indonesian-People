```python
df = pd.read_csv("/content/drive/MyDrive/Data Analysis Project/I. Food Consumption Analysis/konsumsi_pangan_cleanest.csv")
```

```python
df.head()
```

```python
#Calculate total consumption per commodity per year
komoditas_per_tahun = (
    df.groupby(["Tahun", "Komoditas"])["Konsumsi (g/kapita/hari)"]
    .sum()
    .reset_index()
)
```

```python
#Top 5 commodities per year
top_5_komoditas_tahunan = (
    komoditas_per_tahun.sort_values(["Tahun", "Konsumsi (g/kapita/hari)"], ascending=[True, False])
    .groupby("Tahun")
    .head(5)
)
```

```python
print(top_5_komoditas_tahunan)
```

```python
#Highest consumed commodity per day
highest_consumed_commodity = df.sort_values("Konsumsi (g/kapita/hari)", ascending=False).head(5)
```

```python
print(highest_consumed_commodity)
```

```python
#Lowest consumed commodity per day
lowest_consumed_commodity = df.sort_values("Konsumsi (g/kapita/hari)", ascending=True).head(5)
```

```python
print(lowest_consumed_commodity)
```

```python
# overall food consumption (total of all commodities) per capita per day
df.groupby("Tahun")["Konsumsi (g/kapita/hari)"].sum().reset_index()
```

```python
# Classification based on food groups
total_konsumsi = df.groupby(
    ["Tahun", "Kelompok Bahan Pangan", "Gizi"]
)["Konsumsi (g/kapita/hari)"].sum().reset_index()

total_konsumsi = total_konsumsi.sort_values(
    by=["Tahun", "Konsumsi (g/kapita/hari)"],
    ascending=[True, False]
)

top3_per_tahun = total_konsumsi.groupby("Tahun").head(3)
```

```python
print(top3_per_tahun)
```
